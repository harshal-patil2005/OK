```python
#Data Analysis-II

import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
from sklearn.linear_model import LogisticRegression

df = pd.read_csv(r'C:\personal\practice\Social_Network_Ads.csv')
# print(df)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df['Age'],df['EstimatedSalary'])

x_train=x_train.values.reshape(-1,1)
model=LogisticRegression()
model.fit(x_train,y_train)

from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix,f1_score


Y_predicted = model.predict(x_test.values.reshape(-1, 1))

# Confusion Matrix
print(confusion_matrix(y_test, Y_predicted))

# Accuracy
accuracy = accuracy_score(y_test, Y_predicted)
print("Accuracy:", accuracy)

# Precision
precision = precision_score(y_test, Y_predicted, average='weighted')  # use 'micro', 'macro', or 'weighted'
print("Precision:", precision)

recall = recall_score(y_test, Y_predicted, average='weighted')
print("Recall:", recall)
f1 = f1_score(y_test, Y_predicted, average='weighted')  # or 'macro' / 'micro'
print("F1 Score:", f1)
```

```python
# Import necessary libraries
import nltk
import string
from nltk import pos_tag
from collections import Counter
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Download required NLTK data files
nltk.download('punkt')  # Tokenizer models
nltk.download('stopwords')  # List of stopwords
nltk.download('wordnet')  # WordNet lemmatizer data
nltk.download('averaged_perceptron_tagger')  # POS tagger model

# Define the document to analyze
doc = "My name is Gaurav Vilas Kasture and I am 20 years old. I am currently pursuing engineering in computer engineering and I am in the third year. I like computer engineering."

# Tokenize the document into words
tokens = word_tokenize(doc)
print("Tokens:")
print(tokens)

# Perform Part-of-Speech (POS) tagging on the tokens
pos_tags = pos_tag(tokens)
print("\nPart-of-Speech Tags:")
print(pos_tags)

# Remove stopwords from the tokens
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
print("\nTokens after Stopword Removal:")
print(filtered_tokens)

# Remove punctuation from the filtered tokens
filtered_tokens = [word for word in filtered_tokens if word not in string.punctuation]
print("\nTokens after Punctuation Removal:")
print(filtered_tokens)

# Perform stemming on the filtered tokens
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_tokens]
print("\nStemmed Words:")
print(stemmed_words)

# Perform lemmatization on the filtered tokens
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print("\nLemmatized Words:")
print(lemmatized_words)

# Calculate term frequency of the filtered tokens
tf = Counter(filtered_tokens)
print("\nTerm Frequency:")
for word, freq in tf.items():
    print(f"{word}: {freq}")

# TF-IDF Vectorization on a sample corpus
corpus = [
    'This is the first document.',
    'This is the second document.',
    'And the third one.',
    'Is this the first document?',
]

# Initialize the TF-IDF Vectorizer
vectorizer = TfidfVectorizer(use_idf=True)

# Fit the vectorizer on the corpus and transform the corpus into TF-IDF matrix
tfidf_matrix = vectorizer.fit_transform(corpus)

# Retrieve the IDF values for each word in the corpus
idf_values = dict(zip(vectorizer.get_feature_names_out(), vectorizer.idf_))
print("\nIDF Values:")
for word, idf in idf_values.items():
    print(f"{word}: {idf:.4f}")
```

```python
#weblog 
import org.apache.spark.sql.SparkSession
import org.apache.spark.sql.functions._

val spark = SparkSession.builder().appName("WebLog").master("local[*]").getOrCreate()

val baseDF = spark.read.text("weblog.csv")

// Let's take a look at the raw data to understand its structure
baseDF.show(5, false)

// Define a simple schema to extract the needed columns
val logsDF = baseDF.select(
  // Extract host (first word before space)
  split($"value", " ").getItem(0).alias("host"),
  // Extract timestamp (text between square brackets)
  regexp_extract($"value", """\[(\d{2}/\w{3}/\d{4}:\d{2}:\d{2}:\d{2})""", 1).alias("timestamp"),
  // Extract path (text between quotes after method)
  regexp_extract($"value", """\"([^\s]+)""", 1).alias("path"),
  // Extract status code (last value in the line)
  split($"value", " ").getItem(8).cast("int").alias("status")
)

// Drop any rows where critical columns are null
val cleanedDF = logsDF.na.drop()

// Show a quick preview of the cleaned data
cleanedDF.show(5, false)

// Perform basic analysis: Count the number of requests for each status code
cleanedDF.groupBy("status").count().show()

// Perform analysis: Count the number of requests for each host
cleanedDF.groupBy("host").count().show(10)

// Perform analysis: Count the number of requests for each path
cleanedDF.groupBy("path").count().orderBy(desc("count")).show(10)

// Perform a simple count of 404 errors
val notFoundDF = cleanedDF.filter($"status" === 404)
println(s"Total 404 errors: ${notFoundDF.count()}")
notFoundDF.show(10, false)

// Perform daily unique host count
val withDayYear = cleanedDF.withColumn("day", dayofyear(to_timestamp($"timestamp", "dd/MMM/yyyy:HH:mm:ss"))).withColumn("year", year(to_timestamp($"timestamp", "dd/MMM/yyyy:HH:mm:ss")))
val dailyHostsDF = withDayYear.groupBy("day", "year").agg(countDistinct("host").alias("unique_hosts"))
dailyHostsDF.show(5, false)
```
